```python
from PIL import Image
import imagehash
import shelve
import glob
from tqdm import tqdm
```

```python
db = shelve.open('db.shelve', writeback = True)
imagePath = "images/image/ABQE4THPYBFNF6TY.jpeg"
image = Image.open(imagePath)  
hash_value = str(imagehash.dhash(image,4))
# filename = imagePath[imagePath.rfind("/") + 1:]
filename = "1.jpeg"
print(filename)
image.save("images/image/" + filename)

db[hash_value] = db.get(hash_value, []) + [filename]
db.close()
```
